### Importing packages

```python
import face_recognition
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

### Saving known face embeddings

```python
face_dybala = face_recognition.load_image_file('img/dybala.jpg')
enc_dybala = face_recognition.face_encodings(face_dybala)[0]

face_kaka = face_recognition.load_image_file('img/kaka.jpg')
enc_kaka = face_recognition.face_encodings(face_kaka)[0]

face_messi = face_recognition.load_image_file('img/messi.jpg')
enc_messi = face_recognition.face_encodings(face_messi)[0]

face_ronaldinho = face_recognition.load_image_file('img/ronaldinho.jpg')
enc_ronaldinho = face_recognition.face_encodings(face_ronaldinho)[0]

face_ronaldo = face_recognition.load_image_file('img/Ronaldo.jpg')
enc_ronaldo = face_recognition.face_encodings(face_ronaldo)[0]

face_rooney = face_recognition.load_image_file('img/rooney.jpg')
enc_rooney = face_recognition.face_encodings(face_rooney)[0]
```

```python
known_faces = [enc_dybala, enc_kaka, enc_messi, enc_ronaldinho, enc_ronaldo, enc_rooney]
known = ['Dybala', 'Kaka', 'Messi', 'Ronaldinho', 'Ronaldo', 'Rooney']
index = {i: known[i] for i in range(len(known))}
```

### Testing on Unknown Images

#### IMAGE 1

```python
unk_1 = np.array(Image.open('img/unknown/unknown1.jpeg'))

plt.imshow(unk_1)
```

```python
face_unk1 = face_recognition.load_image_file('img/unknown/unknown1.jpeg')
enc_unk1 = face_recognition.face_encodings(face_unk1)[0]

# Compare faces
results = face_recognition.compare_faces(
    known_faces, enc_unk1)
```

```python
flag = False
for i, result in enumerate(results):
    if result==True:
        print('This is {}.'.format(index[i]))
        flag = True
if flag == False:
    print('This is none of the known faces.')
```
